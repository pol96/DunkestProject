```python
import requests as re
from webdriver_manager.chrome import ChromeDriverManager

import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import Select, WebDriverWait

from bs4 import BeautifulSoup as bs

import pandas as pd
from time import sleep
import getpass
import base64 

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
def dunkest_scraping(email, id_players = range(1,10), filter_season = 'Regular Season 20/21'):
    '''
    Recall from Dunkest website all specific statistics game-by-game per player (selected all)
    Variables:
        - email (login email)
        - password (login password)
        - start_id_player (to filter out some players. it will run till exhaustion of ids)
        - filter_season (selected )
    '''
    password = base64.b64encode(getpass.getpass('Password:').encode("utf-8"))
    df = pd.DataFrame()  
    URL = 'https://nba.dunkest.com/it/home'
    # create the initial window
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.maximize_window()  
    # go to the home page
    driver.get(URL)


    sleep(4)

    driver.find_element_by_xpath('//*[@class = "iubenda-cs-opt-group-consent"]').click()
    sleep(1)
    driver.find_element_by_xpath('//*[@class ="dun-btn dun-btn--brand-03 dun-btn--login"]').click()
    sleep(1)

    # enter the email
    driver.find_element_by_xpath('//*[@name ="email"]').send_keys(email)
    driver.find_element_by_xpath('//*[@class ="dun-btn"]').click()  

    sleep(1)

    # enter the password
    driver.find_element_by_xpath('//*[@name ="password"]').send_keys(base64.b64decode(password).decode("utf-8"))
    driver.find_element_by_xpath('/html/body/app-root/div/app-layout-dashboard/app-header/app-login/app-modal/div[1]/div/div/div/div[2]/div[2]/div[2]/button[2]').click()  
    sleep(4)

    # get into player stat page 
    window_before = driver.window_handles[0]
    for i in id_players:
        myurl = "'https://www.dunkest.com/it/giocatore/" +str(i) + "/nikola-jokic?iframe=yes'"
        myurlAction = 'window.open('+myurl+')'
        driver.execute_script(myurlAction)
        window_after = driver.window_handles[1]
        driver.switch_to_window(window_after)
        sleep(1)

        sleep(1)
        try:
            select = Select(driver.find_element_by_xpath('//*[@id="seasonSelect"]'))
            select.select_by_visible_text('Regular Season 21/22')
            sleep(0.66)
            driver.find_element_by_xpath('/html/body/main/div[2]/div/ul/li[2]/span').click()
            sleep(3)
            tab = driver.find_element_by_xpath('//*[@id="gamesTable"]')
            source = tab.get_attribute('innerHTML')
            soup = bs(source)
            data = []
            table_body = soup.find('tbody')

            rows = table_body.find_all('tr')

            for row in rows:
                cols = row.find_all('td')
                cols = [ele.text.strip() for ele in cols]
                data.append([ele for ele in cols if ele]) # Get rid of empty values

            final = pd.DataFrame.from_records(data)
            final['Player'] = driver.find_element_by_xpath('//*[@class = "player-info__firstname"]').text + ' ' +  driver.find_element_by_xpath('//*[@class = "player-info__lastname"]').text
            final['Credits'] = driver.find_element_by_xpath('//*[@class = "player-stats__value"]').text 
            final.columns = ['Data','vs','PDK','MIN','ST','PTS','REB','AST','STL','BLK','BA','FGM','FGA','FG%','3PM','3PA','3P%','FTM','FTA','FT%','OREB','DREB','TOV','PF','+/-','Player','Credits']
            df = df.append(final, ignore_index = True)
            driver.close()
            driver.switch_to_window(window_before)
        except:
            driver.close()
            driver.switch_to_window(window_before)
            pass
    driver.quit()
    return(df)

def credits_scraping(email, season_id,mode, stats_type, start_from, end_to, verbose = True):
    password = base64.b64encode(getpass.getpass('Password:').encode("utf-8"))
    df = pd.DataFrame()  
    URL = 'https://nba.dunkest.com/it/home'
    # create the initial window
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.maximize_window()  
    # go to the home page
    driver.get(URL)


    sleep(4)

    driver.find_element_by_xpath('//*[@class = "iubenda-cs-opt-group-consent"]').click()
    sleep(1)
    driver.find_element_by_xpath('//*[@class ="dun-btn dun-btn--brand-03 dun-btn--login"]').click()
    sleep(1)

    # enter the email
    driver.find_element_by_xpath('//*[@name ="email"]').send_keys(email)
    driver.find_element_by_xpath('//*[@class ="dun-btn"]').click()  

    sleep(1)

    # enter the password
    driver.find_element_by_xpath('//*[@name ="password"]').send_keys(base64.b64decode(password).decode("utf-8"))
    driver.find_element_by_xpath('/html/body/app-root/div/app-layout-dashboard/app-header/app-login/app-modal/div[1]/div/div/div/div[2]/div[2]/div[2]/button[2]').click()  
    sleep(4)

    # get into player stat page 
    window_before = driver.window_handles[0]

    myurl = "'https://www.dunkest.com/it/nba/statistiche/giocatori/tabellone?season_id="+season_id+"&mode="+mode+"&stats_type="+stats_type+"&date_from="+start_from+"&date_to="+end_to+"&teams[]=1&teams[]=2&teams[]=3&teams[]=4&teams[]=5&teams[]=6&teams[]=7&teams[]=8&teams[]=9&teams[]=10&teams[]=11&teams[]=12&teams[]=13&teams[]=14&teams[]=15&teams[]=16&teams[]=17&teams[]=18&teams[]=19&teams[]=20&teams[]=21&teams[]=22&teams[]=23&teams[]=24&teams[]=25&teams[]=26&teams[]=27&teams[]=28&teams[]=29&teams[]=30&positions[]=1&positions[]=2&positions[]=3&player_search=&min_cr=4&max_cr=30&sort_by=pdk&sort_order=desc&iframe=yes'"
    if verbose == True:
        print(myurl)
    myurlAction = 'window.open('+myurl+')'

    driver.execute_script(myurlAction)
    window_after = driver.window_handles[1]
    driver.switch_to_window(window_after)
    sleep(0.66)

    df = pd.DataFrame()
    last_page = int(driver.find_element_by_xpath('//*[@class = "paginationjs-page paginationjs-last J-paginationjs-page"]').text)
    counter = 0
    while counter < last_page:
        try:
            tab = driver.find_element_by_xpath('/html/body/main/div[2]/table')
            source = tab.get_attribute('innerHTML')
            soup = bs(source)
            data = []
            table_body = soup.find('tbody')

            rows = table_body.find_all('tr')

            for row in rows:
                cols = row.find_all('td')
                cols = [ele.text.strip() for ele in cols]
                data.append([ele for ele in cols if ele]) # Get rid of empty values

            final = pd.DataFrame.from_records(data)
            df = df.append(final, ignore_index = True)
            driver.find_element_by_xpath('//*[@class = "paginationjs-next J-paginationjs-next"]').click()
            counter +=1
        except:
            break
    driver.quit()
    df.columns = ['Player','Role','Team','PDK','CR','PLUS','GP','MIN','ST','PTS','REB','AST','STL','BLK','BA','FGM','FGA','FGperc','P3M','P3A','P3perc','FTM','FTA','FTperc','OREB','DREB','TOV','PF','PlusMinus']
    df.drop('PLUS', 1,inplace = True)
    return(df)

def export_df(mail, 
              path = r'C:\Users\ptico\Downloads\\', 
              ranges = range(1,1000), 
              credits = True, 
              season_id = '5',
              mode = 'NBA', 
              stats_type = 'avg', 
              start_from = '2020-12-22', 
              end_to = '2021-05-16'):
    '''
    exporting df in csv format.
    Variables
        - mail (login email)
        - pwd (login password)
        - path (download pointer)
        - name (filename)
        - ranges (id range for selecting downloadable stats for players)
    '''
    if credits:
        name = 'DUN_credits'
        data = credits_scraping(email = mail, 
                                verbose = True, 
                                season_id = season_id, 
                                mode = mode, 
                                stats_type = stats_type, 
                                start_from = start_from,
                                end_to = end_to)
    else:
        name = 'DUN_players'
        data = dunkest_scraping(email = mail, id_players = ranges)
        
    data.to_csv(path+name+'.csv', sep = ',', index = False)
    print(f'download finished for {name}')
```
